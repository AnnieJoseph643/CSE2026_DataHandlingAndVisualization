```python
import pandas as pd
import random
```

```python
Age_Group=["<21", "21-24", "24-27", "27-30","30-33","33-36","36-39","39-42","42-45","45-48","48-51","51-54","54-57","57-60",">60"]
Total_no_of_loans=[random.randint(10,9000) for i in range(15)]
no_of_bad_loans=[random.randint(10,200) for i in range(15)]
no_of_good_loans=[Total_no_of_loans[i]-no_of_bad_loans[i] for i in range(15)]
print(no_of_good_loans)
```

```python
Percentage_of_bad_loans=[((no_of_bad_loans[i]/Total_no_of_loans[i])*100) for i in range(15)]
Percentage_of_good_loans=[((no_of_good_loans[i]/Total_no_of_loans[i])*100) for i in range(15)]
```

```python
data={
    'Age':Age_Group,
    'Total Loans':Total_no_of_loans,
    'Bad Loans':no_of_bad_loans,
    'Good Loans':no_of_good_loans,
    'Percentage Bad Loans':Percentage_of_bad_loans,
    'Percentage Good Loans':Percentage_of_good_loans

}
df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

axes[0, 0].bar(df['Age'], df['Total Loans'])
axes[0, 0].set_title('Total Loans vs Age')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Total Loans')


axes[0, 1].bar(df['Age'], df['Percentage Bad Loans'])
axes[0, 1].set_title('Percentage Bad Loans vs Age')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Percentage Bad Loans')


axes[1, 0].bar(df['Age'], df['Percentage Good Loans'])
axes[1, 0].set_title('Percentage Good Loans vs Age')
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Percentage Good Loans')


axes[1, 1].bar(df['Age'], df['Bad Loans'], label='Bad Loans')
axes[1, 1].bar(df['Age'], df['Good Loans'], bottom=df['Bad Loans'], label='Good Loans')
axes[1, 1].set_title('Bad Loans vs Good Loans')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Number of Loans')
axes[1, 1].legend()


plt.tight_layout()


plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

axes[0, 0].plot(df['Age'], df['Total Loans'], marker='o')
axes[0, 0].set_title('Total Loans vs Age')

axes[0, 1].scatter(df['Age'], df['Percentage Bad Loans'])
axes[0, 1].set_title('Percentage Bad Loans vs Age')

axes[0, 2].scatter(df['Age'], df['Percentage Good Loans'])
axes[0, 2].set_title('Percentage Good Loans vs Age')

df[['Bad Loans', 'Good Loans']].plot(kind='box', ax=axes[1, 0])
axes[1, 0].set_title('Bad Loans vs Good Loans')

total_loans_by_age = df.groupby('Age')['Total Loans'].sum()
axes[1, 1].pie(total_loans_by_age, labels=total_loans_by_age.index, autopct='%1.1f%%')
axes[1, 1].set_title('Total Loans by Age')

axes[1, 2].violinplot(dataset=df['Percentage Bad Loans'], showmeans=False, showmedians=True)
axes[1, 2].set_title('Percentage Bad Loans by Age')
plt.tight_layout()


plt.show()
```
