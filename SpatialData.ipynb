```python
import folium
```

```python
mymap=folium.Map(location=[13.10073,77.59632],
                 width=950,
                 height=550,
                 zoom_starts=12,
                 title='openstreetmap')
mymap
```

```python
folium.Marker([13.10073, 77.59632],
               popup = ' Geeksforgeeks.org ').add_to(mymap)
mymap
```

```python
latitude=12.9716
longitude=77.5946
m=folium.Map(location=[latitude,longitude],zoom_start=12)
m
```

```python
folium.Marker(location=[12.9655881, 77.6036377],popup='Bangalore Palace',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='red',icon='none')).add_to(m)
folium.Marker(location=[12.9715981, 77.5945627],popup='Vidhana Soudha',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='green',prefix='glyphicon',icon='off')).add_to(m)
folium.Marker(location=[12.9660917, 77.5875011],popup='IISC',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='purple',prefix='fa',icon='anchor')).add_to(m)

m
```
